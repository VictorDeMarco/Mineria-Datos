#  Introducción al Análisis de Supervivencia <a id="1"></a><a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>



El **análisis de supervivencia** es un conjunto de métodos estadísticos utilizados para estudiar el tiempo hasta que ocurre un evento de interés. Aunque el término "*supervivencia*" proviene de la investigación clínica donde el evento suele ser la muerte, el reingreso o la recaida esta técnica se aplica a una amplia gama de situaciones donde se analiza la duración hasta la ocurrencia de un evento específico (como por ejemplo la rotura de una pieza).

Podemos verlo como un tipo particular de regresión, en la regresión se predice un valor numérico, pero en el análisis de supervivencia hay características específicas que no ocurren con la regresión.


En **análisis de supervivencia** es especialmente estudiado en el contexto de la salud, donde sirve para contestar preguntas como:

* ¿Cuánto tiempo viven los pacientes después de un diagnóstico de cáncer?
* ¿Cuál es el tiempo hasta la recaída de una enfermedad después del tratamiento?
* ¿Cuánto tiempo transcurre hasta que un paciente con una determinada condición desarrolla una complicación?
* ¿Cuál es la duración de la efectividad de un tratamiento?





# Evento de interés y datos censurados <a id="2"></a><a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>

Para entender el análisis de supervivencia, es crucial conocer los siguientes conceptos:

1.  **Evento de Interés:** Es el suceso específico que estamos estudiando. En salud, puede ser la muerte, la aparición de una enfermedad, la curación, el alta hospitalaria, etc. En industria puede ser la rotura de una pieza, en banca puede ser un impago o una quiebra etc

2.  **Tiempo hasta el Evento:** Es el periodo que transcurre desde un punto de inicio definido (por ejemplo, el diagnóstico, el inicio del tratamiento) hasta que ocurre el evento de interés, como podría ser la recaida, la muerte (o cualquiera de los ejemplos del punto anterior). Esta es la variable principal que se analiza.

3.  **Censura:** Una característica distintiva del análisis de supervivencia es el manejo de datos **censurados**. La censura ocurre cuando no se observa el tiempo completo hasta el evento para algunos individuos. Existen principalmente tres tipos:

    * **Censura a la derecha:** Es el tipo más común. Ocurre cuando el estudio finaliza antes de que el evento ocurra para algunos participantes, o cuando se pierde el seguimiento de un paciente. Sabemos que el tiempo hasta el evento es mayor que el tiempo observado.
        * **Ejemplo:** En un estudio sobre la supervivencia después del tratamiento para una enfermedad, algunos pacientes pueden estar vivos y sin recaída al finalizar el estudio. Su tiempo de supervivencia está censurado a la derecha. No sabemos cuando han recaído o fallecido, pero sabemos a ciencia cierta que será en un momento posterior a la finalización del estudio (y por lo tanto recogida de los datos).

    * **Censura a la izquierda:** Ocurre cuando sabemos que el evento ocurrió antes de un cierto punto en el tiempo, pero no conocemos el momento exacto.
        * **Ejemplo:** Un paciente es diagnosticado con una enfermedad en una revisión médica. Sabemos que la enfermedad comenzó antes de esa revisión, pero no cuándo exactamente.

    * **Censura en un intervalo:** Sabemos que el evento ocurrió dentro de un intervalo de tiempo específico.
        * **Ejemplo:** En un estudio de seguimiento de pacientes, un paciente se encuentra bien en una visita y presenta una recaída en la siguiente. Sabemos que la recaída ocurrió entre esas dos visitas.





#### Ejemplo




4.  **Función de Supervivencia (S(t)):** Representa la probabilidad de que un individuo sobreviva (o no experimente el evento) más allá de un tiempo específico *t*. Matemáticamente, se define como:

    $S(t) = P(T > t)$

    Donde *T* es la variable aleatoria que representa el tiempo hasta el evento. La función de supervivencia comienza en 1 (o 100%) en el tiempo cero y disminuye a medida que transcurre el tiempo.

5.  **Función de Riesgo o Hazard (h(t)):** Representa la probabilidad instantánea de que un individuo experimente el evento en un momento dado *t*, dado que ha sobrevivido hasta ese momento. Se puede interpretar como la tasa a la que ocurren los eventos en un momento específico. Matemáticamente:

    $h(t) \approx P(T < t + \Delta t | T \ge t)$

    Donde $\Delta t$ es un intervalo de tiempo muy pequeño. La función de riesgo proporciona información sobre la ocurrencia del evento, en contraste con la función de supervivencia que describe la ausencia del evento.

6.  **Hazard Ratio (HR):** Es una medida que compara el riesgo de que ocurra el evento entre dos grupos. Por ejemplo, en un ensayo clínico comparando un nuevo tratamiento con un placebo, el HR indicaría cuántas veces mayor es el riesgo de muerte o recaída en un grupo en comparación con el otro.
    * Un HR de 1 indica que no hay diferencia en el riesgo entre los grupos.
    * Un HR mayor que 1 indica un mayor riesgo en el grupo de interés.
    * Un HR menor que 1 indica un menor riesgo en el grupo de interés (es decir, el tratamiento es protector).
        * **Ejemplo:** Un estudio compara la supervivencia de pacientes con cáncer de pulmón tratados con quimioterapia versus radioterapia. Un HR de 0.7 para la quimioterapia en comparación con la radioterapia sugiere que los pacientes tratados con quimioterapia tienen un 30% menos de riesgo de muerte en un momento dado.

### Métodos de Análisis de Supervivencia

Existen varios métodos para analizar datos de supervivencia, cada uno con sus propias suposiciones y aplicaciones:

1.  **Modelo de Kaplan-Meier:** Es un método no paramétrico que estima la función de supervivencia a partir de los datos observados, incluyendo los datos censurados. Genera una curva escalonada que muestra la probabilidad de supervivencia en diferentes puntos del tiempo.
    * **Ejemplo:** Se utiliza para visualizar y comparar las curvas de supervivencia de dos grupos de pacientes con una enfermedad específica, como pacientes que recibieron un tratamiento versus aquellos que no lo recibieron.

2.  **Prueba de Log-Rank:** Es una prueba estadística no paramétrica utilizada para comparar las curvas de supervivencia de dos o más grupos y determinar si existen diferencias significativas entre ellas.
    * **Ejemplo:** Se puede usar para determinar si la curva de supervivencia de pacientes tratados con un nuevo fármaco es significativamente diferente de la curva de supervivencia de pacientes que recibieron el tratamiento estándar.

3.  **Modelo Exponencial:** Es un modelo paramétrico que asume que la función de riesgo es constante a lo largo del tiempo. Aunque simple, esta suposición a menudo no es realista en contextos de salud, ya que el riesgo de muchos eventos (como la muerte) suele cambiar con el tiempo.

4.  **Modelo de Riesgos Proporcionales de Cox:** Es un modelo semi-paramétrico ampliamente utilizado en el análisis de supervivencia. Asume que el hazard ratio entre dos individuos es constante a lo largo del tiempo, pero no especifica la forma de la función de riesgo basal (el riesgo cuando todas las covariables son cero). Permite evaluar el efecto de múltiples covariables (como la edad, el sexo, el estadio de la enfermedad) sobre el tiempo hasta el evento.
    * **Ejemplo:** Se puede utilizar para identificar factores de riesgo independientes para la mortalidad en pacientes con enfermedades cardíacas, ajustando por otras variables como la edad, el índice de masa corporal y la presencia de otras comorbilidades. El modelo proporciona hazard ratios para cada factor, indicando su impacto en el riesgo de muerte.

5.  **Bosques Aleatorios de Supervivencia (Random Survival Forests):** Es un método no paramétrico basado en árboles de decisión que puede manejar relaciones complejas entre las covariables y el tiempo hasta el evento. Es una extensión del algoritmo de Random Forest para datos censurados.



### Evaluación de Modelos de Supervivencia

Para evaluar qué tan bien un modelo de supervivencia predice los resultados, se utilizan varias métricas, incluyendo:

* **Índice de Concordancia (C-index):** Mide la capacidad del modelo para predecir correctamente el orden de los eventos. Un C-index de 0.5 indica una predicción aleatoria, mientras que un valor cercano a 1 indica una predicción perfecta.
* **Área Bajo la Curva ROC Dependiente del Tiempo (Time-dependent AUC):** Evalúa la capacidad del modelo para discriminar entre individuos que experimentarán el evento antes de un tiempo específico y aquellos que no lo harán.
* **Puntuación de Brier Dependiente del Tiempo (Time-dependent Brier Score):** Es una extensión del error cuadrático medio para datos censurados y evalúa tanto la discriminación como la calibración del modelo.



### Conclusión

El análisis de supervivencia es una herramienta esencial en la investigación en salud, ya que permite comprender y modelar el tiempo hasta la ocurrencia de eventos relevantes, teniendo en cuenta la información incompleta proporcionada por los datos censurados. Los diferentes métodos disponibles ofrecen flexibilidad para abordar una variedad de preguntas de investigación y tipos de datos. Comprender los conceptos fundamentales y las técnicas principales del análisis de supervivencia es crucial para interpretar la literatura médica y llevar a cabo investigaciones efectivas en el campo de la salud.

```python
import math

def distribucion_retorno(tir, ano_inicio, ano_final_principal, ano_final_total):
    """
    Parámetros:
      - tir: Tasa Interna de Retorno (por ejemplo, 0.20 para un 20% anual)
      - ano_inicio: Año en el que se empieza a devolver (retornos no nulos antes de este año son 0)
      - ano_final_principal: Año en el que se completa la devolución del principal (suma de principal = 1)
      - ano_final_total: Año en el que se termina la inversión (se devuelven principal + intereses)
    
    Retorna:
      Una lista de flujos (por año, desde el año 0 hasta ano_final_total) con la cantidad devuelta en cada año.
    """
    # Definir el rango de años en los que hay retornos
    anos_retorno = list(range(ano_inicio, ano_final_total + 1))
    
    # Parámetros para la distribución gaussiana:
    # Se elige la media en el centro del período de retornos y se define sigma
    mu = (ano_inicio + ano_final_total) / 2.0
    sigma = (ano_final_total - ano_inicio) / 4.0  # heurística para que la "campana" abarque el período
    
    # Calcular los pesos según una función gaussiana
    pesos = {t: math.exp(-((t - mu) ** 2) / (2 * sigma ** 2)) for t in anos_retorno}
    
    # Separar los años en dos tramos:
    # Tramo principal: desde ano_inicio hasta ano_final_principal (inclusive)
    # Tramo intereses: desde ano_final_principal+1 hasta ano_final_total
    pesos_principal = [pesos[t] for t in anos_retorno if t <= ano_final_principal]
    pesos_intereses = [pesos[t] for t in anos_retorno if t > ano_final_principal]
    
    suma_pesos_principal = sum(pesos_principal)
    suma_pesos_intereses = sum(pesos_intereses) if pesos_intereses else 0

    # Se asume que la suma de las devoluciones del principal debe ser 1.
    # Por ello, para cada año en el tramo principal, el flujo es proporcional al peso.
    # Para el tramo de intereses, se debe determinar el total I (intereses totales) que se va a devolver.
    # La TIR se cumple si:
    #    -1 + [suma_{t en principal} (flujo_t / (1+tir)^t)] + [suma_{t en intereses} (flujo_t / (1+tir)^t)] = 0.
    #
    # Sea A la suma descontada de los flujos del principal y B la suma descontada de la distribución (normalizada)
    # en el tramo de intereses, de modo que:
    #    A = suma_{t en principal} ( (peso_t / suma_pesos_principal) / (1+tir)^t )
    #    B = suma_{t en intereses} ( (peso_t / suma_pesos_intereses) / (1+tir)^t )
    #
    # La ecuación de la TIR queda:
    #    -1 + (1 * A) + (I * B) = 0   =>   I = (1 - A) / B   (si B != 0)
    
    A = sum((pesos[t] / suma_pesos_principal) / ((1 + tir) ** t)
            for t in anos_retorno if t <= ano_final_principal)
    if suma_pesos_intereses > 0:
        B = sum((pesos[t] / suma_pesos_intereses) / ((1 + tir) ** t)
                for t in anos_retorno if t > ano_final_principal)
        I = (1 - A) / B
    else:
        I = 0  # En caso de que no haya tramo de intereses (cuando ano_final_principal == ano_final_total)
    
    # Calcular los flujos anuales según el tramo
    flujos = {}
    for t in anos_retorno:
        if t <= ano_final_principal:
            # Devolución del principal
            flujos[t] = (pesos[t] / suma_pesos_principal) * 1
        else:
            # Devolución de intereses
            flujos[t] = (pesos[t] / suma_pesos_intereses) * I

    # Construir la lista completa de flujos desde el año 0 hasta ano_final_total
    # Asumimos 0 para los años anteriores a ano_inicio
    lista_flujos = [0] * (ano_final_total + 1)
    for t in anos_retorno:
        lista_flujos[t] = flujos[t]
        
    lista_flujos[0] = -1
    
    return lista_flujos


# Parámetros de ejemplo:
tir = 0.12
año_inicio = 4
año_final_principal = 6
año_final_total = 12

flujos_anuales = distribucion_retorno(tir, año_inicio, año_final_principal, año_final_total)
df = pd.DataFrame({"Flujos":flujos_anuales}).round(2)
df    
```

```python
def simulacion_inversion(n_ahorros, n_fondos, tir, ano_inicio_devolucion, 
                         ano_final_principal, ano_final_total):
    """
    Parámetros:
      - n_ahorros: número de años en que el usuario aporta externamente (aporte fijo = 1).
      - n_fondos: número total de años en que se inician fondos.
      - Los parámetros tir, ano_inicio_devolucion, ano_final_principal y ano_final_total se pasan a distribucion_retorno.
      
    Reglas:
      1. Cada año t < n_ahorros se suma 1 a la caja (aporte externo).
      2. Si en el año t se inicia un fondo:
         - Si t < n_ahorros: se financia con ahorro (flujo inicial fijo = –1, y el patrón se escala por 1).
         - Si t ≥ n_ahorros: se financia con caja. Se evalúa la caja disponible (caja del año anterior + aporte externo si aplica).  
             • Si la caja es menor que 1, **no se inicia** el fondo (flujo permanece 0).  
             • Si la caja es ≥ 1, se inicia el fondo con un desembolso igual a  
               \[
               \max(1,\ \text{caja} - 1)
               \]
             y se escala el patrón por ese factor.
      3. Se calcula, año a año:
         - El flujo total (suma de los flujos de todos los fondos activos).
         - La caja, definida como:  
           caja[t] = (caja[t-1] + aporte externo, si t<n_ahorros) – (inversión financiada, si se inicia un fondo con caja) + (flujos de fondos del año).
         - "total": acumulado de los flujos de fondos (sin contar el aporte externo).
      4. Además, se calculan:
         - "aportaciones": el monto invertido (desde ahorro o caja) en el año de inicio de cada fondo.
         - "devoluciones de capital": suma, para todos los fondos, de los flujos recibidos en su índice interno de 1 a (ano_final_principal-1).
         - "rendimientos": suma, para todos los fondos, de los flujos recibidos en su índice interno de ano_final_principal hasta ano_final_total.
         
    Devuelve un DataFrame con filas = años globales (0 a duracion_total-1) y columnas:
      - Una columna por fondo ("Fondo_1", "Fondo_2", …) con sus flujos (ya escalados),
      - "caja", "total", "aportaciones", "devoluciones de capital" y "rendimientos".
    """
    # Patrón para inversión normalizada a 1:
    pattern = distribucion_retorno(tir, ano_inicio_devolucion, ano_final_principal, ano_final_total)
    L = len(pattern)
    
    # Horizonte global: el fondo iniciado en el año (n_fondos-1) tendrá flujos hasta: (n_fondos-1) + L - 1
    duracion_total = (n_fondos - 1) + L
    years = list(range(duracion_total))
    
    # Matriz de flujos: cada columna i corresponde al fondo iniciado en el año i,
    # y se ubica el patrón en las filas t desde i hasta i+L-1 (truncado si excede el horizonte)
    flujos = np.zeros((duracion_total, n_fondos))
    for i in range(n_fondos):
        fund_flow = np.array(pattern)
        start = i
        end = i + L
        if end > duracion_total:
            fund_flow = fund_flow[:duracion_total - i]
        flujos[start: start + len(fund_flow), i] = fund_flow

    # Vector de aportaciones (siempre positivas) registradas en el año de inicio de cada fondo.
    aportaciones = np.zeros(duracion_total)
    
    # Para fondos financiados con ahorro (t < n_ahorros), la aportación es 1 (y el patrón se escala por 1)
    for i in range(n_fondos):
        if i < n_ahorros:
            flujos[i, i] = -1.0
            aportaciones[i] = 1.0

    # Copia de la matriz de flujos para ajustar las aportaciones financiadas con caja
    flujos_ajustados = flujos.copy()
    
    # Inicializamos caja y total (acumulado de flujos de fondos, sin aporte externo)
    caja = np.zeros(duracion_total)
    total_retornos = np.zeros(duracion_total)
    
    # Vectores para acumular devoluciones de capital y rendimientos
    devoluciones = np.zeros(duracion_total)
    rendimientos  = np.zeros(duracion_total)
    
    # Simulación año a año
    for t in range(duracion_total):
        caja_prev = caja[t-1] if t > 0 else 0
        # Aporte externo: si t < n_ahorros, se suma 1 a la caja
        aporte_externo = 1.0 if t < n_ahorros else 0.0
        # Caja disponible antes de financiar un nuevo fondo
        disponible = caja_prev + aporte_externo
        
        # Si en el año t se inicia un fondo financiado con caja (t < n_fondos y t >= n_ahorros)
        if t < n_fondos and t >= n_ahorros:
            if disponible < 1:
                # No se inicia el fondo; la columna correspondiente se queda con 0
                aporte_caja = 0
            else:
                # Se inicia el fondo con un aporte igual al máximo entre 1 y (disponible - 1)
                aporte_caja = max(1, disponible - 1)
                # Nota: esta regla significa que si disponible está entre 1 y 2, se aporta 1;
                # si disponible > 2, se aporta disponible-1.
                
            L_activo = min(L, duracion_total - t)
            flujos_ajustados[t:t+L_activo, t] = np.array(pattern[:L_activo]) * aporte_caja    
        
            aportaciones[t] = aporte_caja
            disponible -= aporte_caja  # se descuenta el aporte financiado de la caja disponible
        
        # El flujo total de fondos en el año t es la suma de los flujos ajustados de ese año
        flujo_fondos = flujos_ajustados[t, :].sum()
        # Actualiza la caja: disponible (ya ajustado) + flujo de fondos
        caja[t] = disponible + flujo_fondos
        # Acumulado total de flujos (retornos de fondos, sin aporte externo)
        total_retornos[t] = (total_retornos[t-1] if t > 0 else 0) + flujo_fondos
    
    # Cálculo de "devoluciones de capital" y "rendimientos":
    # Para cada fondo i (iniciado en el año i), se recorre su vida (t de i hasta min(duracion_total, i+L))
    # y se acumulan los flujos según el índice interno j = t - i:
    #   - Si 1 <= j < ano_final_principal: se suma en devoluciones.
    #   - Si ano_final_principal <= j <= ano_final_total: se suma en rendimientos.
    for i in range(n_fondos):
        for t in range(i, min(duracion_total, i+L)):
            j = t - i
            if j >= 1:
                if j < ano_final_principal:
                    devoluciones[t] += flujos_ajustados[t, i]
                elif j <= ano_final_total:
                    rendimientos[t] += flujos_ajustados[t, i]
    
    # Construcción del DataFrame de resultados:
    # - Una columna por cada fondo ("Fondo_1", "Fondo_2", …) con sus flujos anuales.
    fondos_dict = {f"F._{i+1}": flujos_ajustados[:, i] for i in range(n_fondos)}
    df_result = pd.DataFrame(fondos_dict, index=years)
    df_result["caja"] = caja
    df_result["D Caja"] = devoluciones+rendimientos
    df_result["aportaciones"] = aportaciones
    df_result["devoluciones de capital"] = devoluciones
    df_result["rendimientos"] = rendimientos
    
    return df_result.round(3)

```

```python
# Ejemplo de uso:
# El usuario aporta durante 3 años y se inician fondos durante 5 años.
df_simulacion = simulacion_inversion(n_ahorros=6, n_fondos=12, tir=0.14, 
                                     ano_inicio_devolucion=4, 
                                     ano_final_principal=6, 
                                     ano_final_total=10)
display(df_simulacion)
```
