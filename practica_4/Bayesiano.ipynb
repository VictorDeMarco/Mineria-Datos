````python
def get_frecuencias(df, class_name, laplace=0):
    atributos = df.columns.drop(class_name)
    tablas = []

    for atributo in atributos:
        tabla = pd.crosstab(index=df[class_name], columns=df[atributo])
        if laplace == 1:
            tabla += 1
        tabla.columns = pd.MultiIndex.from_product([[atributo], tabla.columns])
        tablas.append(tabla)
    resultado = pd.concat(tablas, axis=1)
    resultado[("", class_name)] = tablas[0].sum(axis=1)
    return resultado


df_frecs = get_frecuencias(df_play,"play",laplace = 0)
display(df_frecs)

df_frecs = get_frecuencias(df_enfermedad,"Gripe",laplace = 0)
display(df_frecs)

df_frecs = get_frecuencias(df_enfermedad,"Gripe",laplace = 1)
display(df_frecs)
````

###  Tarea 2. Función `get_denominador` <a id="2"></a><a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>

El clasificador bayesiano es, de forma muy simplificada, una tabla que tiene tantas filas como valores diferentes puede tomar la clase y tantas columnas como combinaciones atributo-valor.

En cada casilla de esta tabla hay un cociente entre el valor de la tabla de tabla de frecuencias y el sumatorio a nivel de cada una de las clases.


Para P(valor_atributo | Clase): 

- El denominador para una clase C y un atributo att es el número total de instancias de la clase C, más un ajuste si usamos el suavizado de laplace.

Para P(Clase):
- El denominador es el número total de instancias en el conjunto de datos, más un ajuste si usamos suavizado de laplace.


La función `get_denominador` recibe:
- Un DataFrame de datos.
- El nombre de la columna que actua como clase.
- laplace. Vale 0 si no se aplica laplace y 1 si sí que se aplica.

Devuelve un DataFrame con los valores de los denominadores.




````python
def get_denominador(df,class_name,laplace=0):
    columnas = df.columns.drop(class_name)
    tablas = []

    for col in columnas:
        tabla = pd.crosstab(df[class_name], df[col])
        if laplace == 1:
            tabla += 1
        tabla.columns = pd.MultiIndex.from_product([[col], tabla.columns])
        tablas.append(tabla)
    res = pd.concat(tablas, axis=1)
    res[("", class_name)] = tablas[0].sum(axis=1)
    return res

denominador = get_denominador(df_play,"play",laplace = 0)
display(denominador)

denominador = get_denominador(df_enfermedad,"Gripe",laplace = 0)
display(denominador)

denominador = get_denominador(df_enfermedad,"Gripe",laplace = 1)
display(denominador)
````

````python
def get_denominador(df, class_name, laplace=0):
    clases = df[class_name].value_counts().sort_index()
    atributos = df.columns.drop(class_name)
    new_df = pd.DataFrame(index=atributos, columns=clases.index)

    for atributo in atributos:
        n_valores = df[atributo].nunique()
        new_df.loc[atributo] = clases + laplace * n_valores
    new_df.loc[class_name] = len(df) + laplace * len(clases)
    return new_df.T


denominador = get_denominador(df_play, "play", laplace=0)
display(denominador)

denominador = get_denominador(df_enfermedad, "Gripe", laplace=0)
display(denominador)

denominador = get_denominador(df_enfermedad, "Gripe", laplace=1)
display(denominador)
````

````python

import pandas as pd

def bayesiano(df, class_name, ejemplo, laplace=0):
    atributos = df.columns.drop(class_name)

    frecuencia = get_frecuencias(df, class_name, laplace)

    class_counts = frecuencia.loc[:, ("", class_name)]
    total = frecuencia.loc[:, ("", class_name)].sum()

    prob = class_counts / total

    data = {}
    for atributo in atributos:
        valor = ejemplo[atributo]
        num = frecuencia.loc[:, (atributo, valor)]
        data[(atributo, valor)] = num / class_counts

    data[(class_name, "")] = prob

    cond_df = pd.DataFrame(
        { (atributo, ejemplo[atributo]): data[(atributo, ejemplo[atributo])]
          for atributo in atributos }
    )
    producto = prob * cond_df.prod(axis=1)
    data[("Producto", "")] = producto

    data[("Probs", "")] = producto / producto.sum()

    cols = pd.MultiIndex.from_tuples(list(data.keys()))
    mat  = np.column_stack([data[col] for col in cols])
    resultado = pd.DataFrame(mat, index=class_counts.index, columns=cols)

    if "Sí" in resultado.index:
        nuevo_orden = ["Sí"] + [c for c in resultado.index if c != "Sí"]
        resultado = resultado.reindex(nuevo_orden)
    
    return resultado





bay = bayesiano(df_enfermedad,"Gripe",df_enfermedad_test.iloc[0],laplace = 0)
display(bay)
bay = bayesiano(df_enfermedad,"Gripe",df_enfermedad_test.iloc[1],laplace = 0)
display(bay)

bay = bayesiano(df_enfermedad,"Gripe",df_enfermedad_test.iloc[0],laplace = 1)
display(bay)
bay = bayesiano(df_enfermedad,"Gripe",df_enfermedad_test.iloc[1],laplace = 1)
display(bay)

````
